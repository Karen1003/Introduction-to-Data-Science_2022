## import所需套件

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
import nltk
import re
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))
```

## 讀取資料

```python
df=pd.read_csv('train.csv')
print(df.head())
```

## 資料清理
### 正則化、移除停用詞、SnowballStemmer轉換字詞變化

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
```

```python
df['tweet']=df['tweet'].apply(clean_text)

df.head()
```

## 特徵分析

```python
#文字雲
from wordcloud import WordCloud, STOPWORDS
stopwords = STOPWORDS.add('amp')

#紀錄所有資料及區分class
all_words = ' '.join(df.tweet.values)
hateful = ' '.join(df[df["class"]==0].tweet.values)
offensive=' '.join(df[df["class"]==1].tweet.values)
clean=' '.join(df[df["class"]==2].tweet.values)


cloud1 = WordCloud(width=400, height=400, background_color='white', stopwords=stopwords).generate(all_words)
plt.subplot(121)
plt.imshow(cloud1, interpolation="bilinear")
plt.axis("off")
plt.title('All tweets', size=20)

cloud2 = WordCloud(width=400, height=400, background_color='white', stopwords=stopwords).generate(hateful)
plt.subplot(122)
plt.imshow(cloud2, interpolation="bilinear")
plt.axis("off")
plt.title('Hateful tweets', size=20)
plt.show()

cloud3 = WordCloud(width=400, height=400, background_color='white', stopwords=stopwords).generate(offensive)
plt.subplot(122)
plt.imshow(cloud3, interpolation="bilinear")
plt.axis("off")
plt.title('Offensive tweets', size=20)
plt.show()

cloud4 = WordCloud(width=400, height=400, background_color='white', stopwords=stopwords).generate(clean)
plt.subplot(122)
plt.imshow(cloud4, interpolation="bilinear")
plt.axis("off")
plt.title('Clean tweets', size=20)
plt.show()
```

```python
#目標欄位分布
ax = sns.countplot(x=df['class'],palette=sns.cubehelix_palette(2))
plt.bar_label(ax.containers[0])
plt.title("class Distribution", fontsize=18,color="red");
```

## 切割訓練資料自行測試

```python
X=df['tweet']
Y=df['class']

#特徵萃取
cv = CountVectorizer()
X=cv.fit_transform(X).toarray()

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
#模型訓練及測試
clf=DecisionTreeClassifier()
bag = BaggingClassifier(base_estimator= clf)
bag.fit(X_train,y_train)

print("BaggingClassifier Accuracy :",bag.score(X_test, y_test))
print(classification_report(y_test,  bag.predict(X_test)))
```

## 利用所有訓練資料建立模型並輸出上傳檔案

```python
#建立模型
clf=DecisionTreeClassifier()
bag = BaggingClassifier(base_estimator= clf)
bag.fit(X,Y)
```

```python
#預測結果
pred = bag.predict(df_test)
pred
```
