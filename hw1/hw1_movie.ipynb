# 第二題

```python
data=pd.read_csv('IMDB-Movie-Data.csv')
#Revenue為0的列刪除並重新排列index
data=data.dropna(how="any",subset=['Revenue (Millions)'],inplace=False).reset_index() 

actor_list=[]
actor_list_count=[] #紀錄該部電影有幾位演員
#將每一列actor取出並將所有演員extend到actor_list中
for i in range(0,len(data)):
    actor=data['Actors'][i].split('|')
    #去除前面空格
    for j in range(0,len(actor)):
        actor[j]=actor[j].strip()
    actor_list.extend(actor)
    actor_list_count.append(len(actor))
#pd.Series(actor_list)

revenue_list=[] #按照actor把revenue依序排出來
for i in range(0,len(data)):
    for j in range(0,actor_list_count[i]):
        revenue_list.append(data['Revenue (Millions)'][i])
        
# 將actor和對應的revenue製成一個dataframe
actor_revenue=pd.DataFrame({'Actor': actor_list,'Revenue': revenue_list})

#把個別actor的revenue計算平均並且由大排到小
actor_revenue=actor_revenue.groupby('Actor').mean().sort_values(['Revenue'],ascending=False)

#第0筆為average revenue最大的演員。記錄並列排名到第幾列
average_revenue=actor_revenue['Revenue'][0]
for i in range(0,len(actor_revenue)):
    if actor_revenue['Revenue'][i]==average_revenue:
        result=i
#印出結果        
print(actor_revenue[0:result+1])
```

# 第五題

```python
data=pd.read_csv('IMDB-Movie-Data.csv')
#Revenue為0的列刪除並重新排列index
data=data.dropna(how="any",subset=['Revenue (Millions)'],inplace=False).reset_index() 

actor_list=[]
actor_list_count=[] #紀錄該部電影有幾位演員
#將每一列actor取出並將所有演員extend到actor_list中
for i in range(0,len(data)):
    actor=data['Actors'][i].split('|')
    #去除前面空格
    for j in range(0,len(actor)):
        actor[j]=actor[j].strip()
    actor_list.extend(actor)
    actor_list_count.append(len(actor))
    
genre_list=[] #按照actor把genre依序排出來
for i in range(0,len(data)):
    for j in range(0,actor_list_count[i]):
        genre_list.append(data['Genre'][i])

        
# 將actor和對應的genre製成一個dataframe
actor_genre=pd.DataFrame({'Actor': actor_list,'Genre': genre_list})

actor_genre=actor_genre.sort_values(by=['Actor']).reset_index(drop=True)


genre_list=[]
genre_list_count=[] #紀錄該部電影包含幾種類型
#將每一列genre取出並將所有genre extend到genre_list中
for i in range(0,len(actor_genre)):
    genre=actor_genre['Genre'][i].split('|')
    genre_list.extend(genre)
    genre_list_count.append(len(genre))

    
actor_list=[] #按照genre把actor依序排出來
for i in range(0,len(actor_genre)):
    for j in range(0,genre_list_count[i]):
        actor_list.append(actor_genre['Actor'][i])
        
        
df=pd.DataFrame({'Actor': actor_list,'Genre': genre_list})

#刪掉重複的資料
df=df.drop_duplicates().reset_index(drop=True)

#類型前兩高的演員
df['Actor'].value_counts().head(6)
```

# 第六題

```python
data=pd.read_csv('IMDB-Movie-Data.csv')
data=data.fillna(0) #空白處補0

actor_list=[]
actor_list_count=[] #紀錄該部電影有幾位演員
#將每一列actor取出並將所有演員extend到actor_list中
for i in range(0,len(data)):
    actor=data['Actors'][i].split('|')
    for j in range(0,len(actor)):
        actor[j]=actor[j].strip()
    actor_list.extend(actor)
    actor_list_count.append(len(actor))


year_list=[] #按照actor把year依序排出來
for i in range(0,len(data)):
    for j in range(0,actor_list_count[i]):
        year_list.append(data['Year'][i])
        
# 將actor和對應的year製成一個dataframe
actor_year=pd.DataFrame({'Actor': actor_list,'Gap Year': year_list})

#按照actor分群將最大年分-最小年分，並且按照gap year由大到小排序
actor_year=(actor_year.groupby('Actor').max()-actor_year.groupby('Actor').min()).sort_values(['Gap Year'],ascending=False)
actor_year=actor_year.reset_index() 
actor_year

#第0筆為gap year最大，index紀錄並排到第幾筆
max_gap=actor_year['Gap Year'][0]
for i in range(0,len(actor_year)):
    if actor_year['Gap Year'][i]==max_gap:
        index=i
        
print(actor_year[0:index+1])
```
