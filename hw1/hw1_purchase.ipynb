### In-Game Purchase Data Analysis

#### Background 

In this homework, you are ask to deal with a task of analyzing an "in-game purchase" dataset. In this game, players are able to to purchase optional items that enhance their playing experience. Now your task is to generate a report that breaks down the game's purchasing data into meaningful insights.

-----

#### Observable Trends

* There are 1163 active players. The vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.79%) with secondary groups falling between 15-19 (18.58%) and 25-29 (13.37%). 

* The age group that spends the most money is the 20-24 with 1,114.06 dollars as total purchase value and an average purchase of 4.32. In contrast, the demographic group that has the highest average purchase is the 35-39 with 4.76 and a total purchase value of 147.67. 
-----


```python
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Age Demographics

* Categorize players using the age groups. Hint: use pd.cut()
  * each group is for 4 years (i.e. &lt;10, 10-14, 15-19, etc.)
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* round the percentage column to two decimal points
* Display Age Demographics Table

```python
purchase_data_2=purchase_data.drop_duplicates(subset=['SN'],ignore_index=True) #把player重複的資料刪除

age_count=purchase_data_2['Age'].value_counts().sort_index(ascending=True).to_frame() #按照年紀小到大計算各有幾人並轉成dataframe

player=[]
player.append(age_count['Age'][0:3].sum()) #<10的人數 

#依序計算累計人數，每次跳五格
for i in range(3,len(age_count),5):
    player.append(age_count['Age'][i:i+5].sum())
    
#最後多出一格45歲要加回40-44歲
player[(len(player)-2)]=player[(len(player)-2)]+player[len(player)-1]
player.remove(player[len(player)-1])


percentage=[]
total=len(purchase_data_2)

#依序計算百分比，並使用正規表達式表示百分比
for i in range(len(player)):
    percentage.append('{:.2%}'.format(player[i]/total))
    #percentage.append(round((count[i]/total)*100,2))
    
pd.DataFrame({'Percentage of Players':percentage,'Total Count':player},index=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'])
```

## Purchasing Analysis (Age)

* The below each broken into groups of 4 years (i.e. &lt;10, 10-14, 15-19, etc.)
  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value
  * Average Purchase Total per Person by Age Group
* Bin the purchase_data data frame by age
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
* Create a summary data frame to hold the results
* Display the summary data frame

```python
#Average Purchase Total per Person=Total Purchase Value/上題的Total Count

age_count=purchase_data['Age'].value_counts().sort_index(ascending=True).to_frame() #按照年紀小到大計算各有幾人並轉成dataframe


count=[]
count.append(age_count['Age'][0:3].sum()) #<10的人數 

#依序計算累計人數，每次跳五格
for i in range(3,len(age_count),5):
    count.append(age_count['Age'][i:i+5].sum())
    
#最後多出一格45歲要加回40-44歲
count[(len(count)-2)]=count[(len(count)-2)]+count[len(count)-1]
count.remove(count[len(count)-1])



#total purchase value
price_count=purchase_data.groupby('Age').sum()

price=[]
price.append(price_count['Price'][0:3].sum()) #<10的人數 

#依序計算累計價格，每次跳五格
for i in range(3,len(price_count),5):
    price.append(price_count['Price'][i:i+5].sum())
    
#最後多出一格45歲要加回40-44歲
price[(len(price)-2)]=price[(len(price)-2)]+price[len(price)-1]
price.remove(price[len(price)-1])


#Average Purchase Price
price_a=[]
for i in range(len(price)):
    price_a.append(round(price[i]/count[i],2))
    
    
#Average Purchase Total per Person
price_a_p=[]
for i in range(len(price)):
    price_a_p.append(round(price[i]/player[i],2))

pd.DataFrame({'Purchase Count':count,'Average Purchase Price':price_a,'Total Purchase Value':price,'Average Purchase Total per Person':price_a_p},index=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'])
```

## Top Spenders

* Identify the the top 5 spenders in the game by total purchase value (top-5 players with highest total purchase value), then list (in a table):
  * SN(screen name)
  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value
* Run basic calculations to obtain the results in the table below
* Create a summary data frame to hold the results
* Sort the total purchase value column in descending order
* Display a preview of the summary data frame

```python
#總消費金額排前五的player
price_rank=purchase_data.groupby('SN').sum().sort_values(['Price'],ascending=False).reset_index()
SN=price_rank['SN'][0:5]
Total_Purchase_Value=price_rank['Price'][0:5]

#各玩家的消費次數，對應到消費金額排前五的選手紀錄下消費次數到Purchase_Count
count=purchase_data['SN'].value_counts().rename_axis('SN').reset_index(name='counts')
Purchase_Count=[]
for j in range(5):
    for i in range(len(count)):
        if count['SN'][i]==SN[j]:
            Purchase_Count.append(count['counts'][i])
#Average Purchase Price
average_purchase=round(Total_Purchase_Value/Purchase_Count,2)

df=pd.DataFrame({'Purchase Count':Purchase_Count,'Average Purchase Price':average_purchase,'Total Purchase Value':Total_Purchase_Value})
df.set_index([SN])
```

## Most Popular Items

* Top 5 most popular items by purchase count:
  * Item ID
  * Item Name
  * Purchase Count
  * Item Price
  * Total Purchase Value
* Retrieve the Item ID, Item Name, and Item Price columns
* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
* Create a summary data frame to hold the results
* Sort the purchase count column in descending order
* Display a preview of the summary data frame

```python
#以Item ID計算被購買次數最多的前五名，取出Item ID和購買次數
purchase_count=purchase_data['Item ID'].value_counts().rename_axis('Item ID').reset_index(name='Purchase Count')
ID=purchase_count['Item ID'][0:5]
count=purchase_count['Purchase Count'][0:5]

#找出Item ID對應的商品名稱和價格
Item_Name=[]
Item_Price=[]
for i in range(len(ID)):
    Item_Name.append(purchase_data[purchase_data['Item ID']==ID[i]].reset_index(drop=True)['Item Name'][0])
    Item_Price.append(purchase_data[purchase_data['Item ID']==ID[i]].reset_index(drop=True)['Price'][0])

total_value=purchase_data.groupby('Item ID').sum().reset_index()
Total_Purchase_Value=[]
for i in range(len(ID)):
    Total_Purchase_Value.append(total_value[total_value['Item ID']==ID[i]].reset_index(drop=True)['Price'][0])

df=pd.DataFrame({'Item Name':Item_Name,'Purchase Count':count,'Item Price':Item_Price,'Total Purchase Value':Total_Purchase_Value})

df.set_index([ID])
```

## Most Profitable Items

* Top 5 most profitable items by total purchase value:
  * Item ID
  * Item Name
  * Purchase Count
  * Item Price
  * Total Purchase Value
* Sort the above table by total purchase value in descending order
* Display a preview of the data frame

```python
#以Item ID計算被購買總額最多的前五名，取出Item ID和購買金額
purchase_count=purchase_data.groupby('Item ID').sum().sort_values(['Price'],ascending=False).reset_index()
purchase_count=purchase_count.drop(['Purchase ID','Age'],axis=1)
ID=purchase_count['Item ID'][0:5]
price=purchase_count['Price'][0:5]

#groupby Item ID的購買次數
count=purchase_data['Item ID'].value_counts().rename_axis('Item ID').reset_index(name='counts')
#找出Item ID對應的商品名稱、購買次數和價格
Item_Name=[]
Purchase_Count=[]
Item_Price=[]
for i in range(len(ID)):
    Item_Name.append(purchase_data[purchase_data['Item ID']==ID[i]].reset_index(drop=True)['Item Name'][0])
    Purchase_Count.append(count[count['Item ID']==ID[i]].reset_index(drop=True)['counts'][0])
    Item_Price.append(purchase_data[purchase_data['Item ID']==ID[i]].reset_index(drop=True)['Price'][0])

df=pd.DataFrame({'Item Name':Item_Name,'Purchase Count':Purchase_Count,'Item Price':Item_Price,'Total Purchase Value':price})
df.set_index([ID])
```
