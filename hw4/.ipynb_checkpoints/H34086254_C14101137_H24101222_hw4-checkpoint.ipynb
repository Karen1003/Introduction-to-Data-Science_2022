```python
import numpy as np
import pandas as pd

# For visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

# 讀取資料

```python
#讀取train.csv並確認資料結構
df = pd.read_csv('train.csv', delimiter=',')
df.shape
```

# 資料內容觀察

```python
#檢查有無缺失值
df.isnull().sum()
```

```python
#刪除與結果無關聯的欄位
df = df.drop(["RowNumber", "CustomerId", "Surname"], axis = 1)
df.head()
```

```python
#非連續型資料分布情形
fig, axarr = plt.subplots(2, 2, figsize=(20, 12))
sns.countplot(x='Geography', hue = 'Exited',data = df, ax=axarr[0][0])
sns.countplot(x='Gender', hue = 'Exited',data = df, ax=axarr[0][1])
sns.countplot(x='HasCrCard', hue = 'Exited',data = df, ax=axarr[1][0])
sns.countplot(x='IsActiveMember', hue = 'Exited',data = df, ax=axarr[1][1])
```

```python
#連續型資料分布情形
fig, axarr = plt.subplots(3, 2, figsize=(20, 12))
sns.boxplot(y='CreditScore',x = 'Exited', hue = 'Exited',data = df, ax=axarr[0][0])
sns.boxplot(y='Age',x = 'Exited', hue = 'Exited',data = df , ax=axarr[0][1])
sns.boxplot(y='Tenure',x = 'Exited', hue = 'Exited',data = df, ax=axarr[1][0])
sns.boxplot(y='Balance',x = 'Exited', hue = 'Exited',data = df, ax=axarr[1][1])
sns.boxplot(y='NumOfProducts',x = 'Exited', hue = 'Exited',data = df, ax=axarr[2][0])
sns.boxplot(y='EstimatedSalary',x = 'Exited', hue = 'Exited',data = df, ax=axarr[2][1])
```

# 特徵工程與資料前處理

```python
#創建新特徵
df['BalanceSalaryRatio'] = df.Balance/df.EstimatedSalary

df['TenureByAge'] = df.Tenure/(df.Age)

df['CreditScoreGivenAge'] = df.CreditScore/(df.Age)
```

```python
#確認現在所有的欄位
df.head()
```

```python
#針對Gender和Geography欄位進行轉換
df_dum = pd.get_dummies(df)
df_dum.head()
```

```python
#將連續型與類別型資料欄位名稱分別記錄下來，以利後續進行資料整理
continuous_vars = ['CreditScore',  'Age', 'Tenure', 'Balance','NumOfProducts', 'EstimatedSalary', 'BalanceSalaryRatio',
                   'TenureByAge','CreditScoreGivenAge']
cat_vars = ['HasCrCard', 'IsActiveMember','Geography_France','Geography_Germany','Geography_Spain','Gender_Female','Gender_Male']
```

```python
# 所有連續型資料
df_dum[continuous_vars].head()
```

```python
import sklearn.preprocessing as sp
```

```python
#連續型資料進行標準化
fs = sp.StandardScaler().fit(df_dum[continuous_vars])
result_std = fs.transform(df_dum[continuous_vars])
```

```python
df_std = pd.DataFrame(result_std,columns=continuous_vars)
df_std.head()
```

```python
#所有類別型資料
df_dum[cat_vars].head()
```

```python
#將連續型與類別型資料重新合併，連續型在前，類別型在後
final_df = pd.concat((df_std,df_dum[cat_vars]),axis= 1)
final_df.head()
```

# 模型測試

## 自行切割train資料進行模型測試

```python
data_y = df["Exited"]
data_X = final_df
```

```python
#資料切割
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data_X,data_y,test_size=0.2)
```

```python
#模型訓練與測試
from sklearn import tree
from sklearn.ensemble import BaggingClassifier
clf = tree.DecisionTreeClassifier(criterion= 'entropy',max_depth= 8)
bag = BaggingClassifier(base_estimator= clf,n_estimators= 100,max_features= 0.7,n_jobs= -1)
bag.fit(X_train,y_train)

from sklearn.metrics import classification_report
print(classification_report(y_test,  bag.predict(X_test)))
print(bag.score(X_test,y_test))
```

# 模型訓練與上傳

## 利用train.csv進行模型訓練

```python
#模型訓練
model = tree.DecisionTreeClassifier(criterion= 'entropy',max_depth= 8)
model = BaggingClassifier(base_estimator= model,n_estimators= 100,max_features= 0.3,n_jobs= -1)
model.fit(data_X,data_y)
```

## 利用test.csv進行預測，資料處理流程同train.csv

```python
test_data = pd.read_csv("test.csv")
test_data.head()
```

```python
td = test_data.drop(["RowNumber", "CustomerId", "Surname"], axis = 1)
td.head()
```

```python
td['BalanceSalaryRatio'] = td.Balance/td.EstimatedSalary
td['TenureByAge'] = td.Tenure/(td.Age)
td['CreditScoreGivenAge'] = td.CreditScore/(td.Age)
```

```python
td_dum = pd.get_dummies(td)
td_dum.head()
```

```python
td_dum[continuous_vars].head()
```

```python
result_std_td = fs.transform(td_dum[continuous_vars])
```

```python
td_std = pd.DataFrame(result_std_td,columns=continuous_vars)
td_std.head()
```

```python
final_test_df = pd.concat((td_std,td_dum[cat_vars]),axis= 1)
final_test_df.head()
```

```python
#生成預測結果
test_result = model.predict(final_test_df)
test_result
```

# 輸出成csv

```python
#生成dataframe
column1 = pd.DataFrame(test_result)
column1.head()
```
